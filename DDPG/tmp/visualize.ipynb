```python
# load test history json
import json
import matplotlib.pyplot as plt
import numpy as np

with open('test_history.json') as f:
    test_history = json.load(f)

# load episode duration json
with open('episode_data_dict.json') as f:
    episode_duration = json.load(f)
rounded_dur = [sum(episode_duration['durations'][i:i+15])/15 for i in range(0, len(episode_duration['durations']), 15)]
# plot episode duration
plt.figure(figsize=(10, 5))
plt.plot(rounded_dur)
plt.xlabel('episode')
plt.ylabel('duration')
plt.title('episode duration')

ROUND_EVERY = 10
# round every 10 episodes
test_history = np.array(test_history)
rounded = [sum(test_history[i:i+ROUND_EVERY])/ROUND_EVERY for i in range(0, len(test_history), ROUND_EVERY)]
# plot test history
plt.figure(figsize=(10, 5))
plt.plot(rounded)
plt.xlabel('episode')
plt.ylabel('steps')
plt.title('test history')
plt.show()

```
